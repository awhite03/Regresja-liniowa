```python
import numpy as np
data_np = np.array(data_encoder_video, dtype=np.int16)
target_np = np.array(target, dtype=np.int16)

print(type(data_np))
print(type(target_np))

print(data_np.shape)
print(target_np.shape)
```

```python
print("---Mean---")
print(data_np.mean(axis=0))
print("---std---")
print(data_np.std(axis=0))
```

```python
#Data standardisation with StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_np)
```

```python
print("---Mean---")
print(scaled_data.mean(axis=0))
print("---std---")
print(scaled_data.std(axis=0))
```

```python
from sklearn.model_selection import train_test_split

video_train_data, video_test_data, \
video_train_target, video_test_target = \
train_test_split(scaled_data,target, test_size=0.1, random_state=10)
```

```python
print("Training dataset:")
print("video_train_data:", video_train_data.shape)
print("video_train_target:", video_train_target.shape)
```

```python
print("Training dataset:")
print("video_test_data:", video_test_data.shape)
print("video_test_target:", video_test_target.shape)
```

```python
#Training LinearRegression model
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(video_train_data, video_train_target)
```

```python
#Model evaluation
```

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error of a learned model: %.2f" % 
      mean_squared_error(video_test_target, linear_regression.predict(video_test_data)))
```

```python
from sklearn.metrics import r2_score
print('Variance score: %.2f' % r2_score(video_test_target, linear_regression.predict(video_test_data)))
```

```python
#Model evaluation with cross-validation
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(LinearRegression(), scaled_data, target, cv=4)
print(scores)
```
